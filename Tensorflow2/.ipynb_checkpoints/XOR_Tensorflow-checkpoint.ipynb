```python
# Tensorflow
import tensorflow as tf

# Numpy
import numpy as np
```

```python
# Hidden Layer가 있는 신경망 모델 구성, activation은 sigmoid
model = tf.keras.models.Sequential([
  tf.keras.layers.Dense(4, activation='sigmoid'),
  tf.keras.layers.Dense(4, activation='sigmoid'),
  tf.keras.layers.Dense(1, activation='sigmoid'),
])
```

```python
# XOR dataset
X = tf.constant([[0, 0], [1, 0], [0, 1], [1, 1]])
Y = tf.constant([0, 1, 1, 0])
```

```python
# (None, 1) 꼴의 데이터를 Input으로 받을 것이다.
model.build((None, 2))
# 모델 요약 보기
model.summary()

# compile (Loss, optimizer 설정)
adam = tf.keras.optimizers.Adam(learning_rate=0.01)
model.compile(optimizer=adam,
              loss='binary_crossentropy',
              )
```

```python
model.fit(X, Y, epochs=500)
```

```python
print(model(X))
# 0, 1, 1, 0에 유사한 값이 나오는 것을 볼 수 있음
```

비선형 문제를 Hidden Layer를 이용한 신경망 구조로 해결할 수 있게 되었다.
